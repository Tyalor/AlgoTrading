```python
import yfinance as yf
import pandas as pd
from statsforecast import StatsForecast
from statsforecast.models import AutoARIMA, AutoETS, HistoricAverage, Naive, RandomWalkWithDrift, SeasonalNaive
import matplotlib.pyplot as plt

# Load your data
stock_symbols = 'QQQ'
start_date = '2023-06-01'
end_date = '2024-01-01'
df = yf.download(stock_symbols, start=start_date, end=end_date)

# Prepare the data for historical prices (date and 'Close' price)
df_historical = df.reset_index()[['Date', 'Close']]  # Select the 'Close' column for historical prices
df_historical.columns = ['ds', 'y']  # Rename columns to 'ds' and 'y'
df_historical['unique_id'] = "1"
df_historical.to_csv("qqq-historical-data.csv")  # Save the historical data to a CSV file

# Plotting the historical data
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Close'], label='Historical Prices')
plt.title(f'{stock_symbols} Stock Prices from {start_date} to {end_date}')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()

models = [
    AutoETS(season_length=24),
    HistoricAverage(), 
    Naive(), 
    RandomWalkWithDrift(), 
    SeasonalNaive(season_length=24)
]

# Create a StatsForecast object and specify the models
sf = StatsForecast(
    models=models, 
    freq='D'
)

sf.fit(df_historical)
forecast_df = sf.forecast(h=60, level=[95])
forecast_df

# average_forecast = []
# for index, row in forecast_df.iterrows():
#     average_forecast.append((row["AutoETS-lo-95"] + row["AutoETS-hi-95"]) / 2)
# forecast_df['Average_ETSprediction'] = average_forecast

# average_forecast

# Plotting the forecast
plt.figure(figsize=(10, 6))
plt.plot(df_historical['ds'], df_historical['y'], label='Historical')
plt.plot(forecast_df['ds'], forecast_df['SeasonalNaive'], label='Forecast (Average)', color='orange')
plt.title(f'{stock_symbols} Stock Price Forecast (Average)')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```
